Имеется AI по типу ChatGPT. Выпустили обновление, которое должно улучшить выдачу ответов и снизить кол-во нажатий на 'generate answer again' до среднего значения равного 0.2 . Необходимо узнать, действительно ли оно лучше.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.stats.api as sms
from tqdm import tqdm
%matplotlib inline

sns.set(style="ticks", context="talk")
plt.style.use("dark_background")
```

```python
size = 10000000 # Кол-во юзеров данного сервиси
user_id = np.random.choice(size, size, replace=False) # Сгенерирую id пользователей
current_mean = 0.25 # Задам текущее среднее значение
shape = 2 # Форму распределение
scale = current_mean / shape # Scale, чтобы среднее было равно current_mean

gen_again = np.random.gamma(shape, scale=scale, size=size)
data = pd.DataFrame({'user_id': user_id,
                     'gen_again': gen_again})
data.head()
```

```python
data.gen_again.mean()
```

```python
print(data.gen_again.describe())
plt.figure(figsize=(20,7))
sns.histplot(data=data.gen_again, bins=50)
```

```python
# Рассчитаю размер выборки
effect_size = sms.proportion_effectsize(0.20, 0.25)
sample_size = round(sms.NormalIndPower().solve_power(effect_size, power=0.9, alpha=0.01, ratio=1)) * 2
```

```python
# Сплит система. АА тест.
def split(uid):
    group = hash(str(uid) + 'hello') % 2
    return 'test' if group else 'control'


t_data = data.sample(sample_size) # Сделаю сэмпл, чтобы потом разбить на группы.
t_data['group'] = t_data.user_id.apply(split)
control_group = t_data.query('group == "control"')
test_group = t_data.query('group == "test"')
stats.ttest_ind(control_group.gen_again, test_group.gen_again, equal_var=False)
```

```python
fit, axes = plt.subplots(nrows=2, figsize=(10,10))
plt.subplots_adjust(hspace=0.8, top=0.9, bottom=0.1)
sns.histplot(data=control_group.gen_again, ax=axes[0])
axes[0].set_title('Control Group')
sns.histplot(data=test_group.gen_again, ax=axes[1])
axes[0].set_title('Test Group')
```

Сплит система работает корректно. Проводим A/B test.


```python
test_group.gen_again = test_group.gen_again.apply(lambda x: x * np.random.uniform(0.55, 1.05)) # То есть в среднем значение должно стать 0.2
test_group.head()
```
